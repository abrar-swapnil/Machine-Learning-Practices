# Real Dataset Load

```python
import pandas as pd
from sklearn.datasets import load_diabetes

# ১. Real diabetes data load korlam
diabetes = load_diabetes(as_frame=True)
df_diabetes = diabetes.frame

# ২. Dataset-er prothom 5 ti row dekhi
print(df_diabetes.head())
```

# Convert Text to Number

```python
# Dummy context-er jonno ekti City column jog korlam (Text data)
df_diabetes['City'] = ['Dhaka', 'Chittagong', 'Sylhet', 'Dhaka', 'Sylhet'] * 88 + ['Dhaka', 'Chittagong']

# Text column-ke one-hot encoding korar standard Pandas function:
df_encoded = pd.get_dummies(df_diabetes, columns=['City'], drop_first=True)

print("\nEncoding korar por nuthun columns gulo dekho:")
print(df_encoded.head())
```

# Target Split,Training & R^2 Score Test

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

# ১. X and y separation
# Target column bad diye baki shob column holo X
X_real = df_encoded.drop(columns=['target'])
y_real = df_encoded['target']

# ২. Train-Test Split (80% Train, 20% Test)
X_train_r, X_test_r, y_train_r, y_test_r = train_test_split(X_real, y_real, test_size=0.2, random_state=42)

# ৩. Model setup & training
real_model = LinearRegression()
real_model.fit(X_train_r, y_train_r)

# ৪. Prediction & R2 Score Calculation
real_predictions = real_model.predict(X_test_r)
real_r2 = r2_score(y_test_r, real_predictions)

print(f"\nReal Diabetes Dataset-e Model-er R2 Score: {real_r2:.4f}")
```

# Tuning solution

```python
from sklearn.tree import DecisionTreeRegressor # <- Regressor tree imported
from sklearn.metrics import r2_score

# ১. Model Object setup
tree_reg = DecisionTreeRegressor(random_state=42)

# ২. Model Training phase
tree_reg.fit(X_train_r, y_train_r)

# ৩. Testing Exam & Scoring
tree_pred = tree_reg.predict(X_test_r)
tree_r2 = r2_score(y_test_r, tree_pred)

print(f"Decision Tree Regressor er R2 Score: {tree_r2:.4f}")
```

```python
# tree-r depth limit kore dilam max_depth=3 diye
tuned_tree = DecisionTreeRegressor(max_depth=3, random_state=42)

# Re-train 
tuned_tree.fit(X_train_r, y_train_r)

# Re-test and Evaluate
tuned_pred = tuned_tree.predict(X_test_r)
tuned_r2 = r2_score(y_test_r, tuned_pred)

print(f"Tuned Decision Tree er R2 Score: {tuned_r2:.4f}")
```

Limiting the tree depth via the ⁠max_depth=3⁠ parameter forced the model to ignore micro-level noise and prioritize the top 3 structural data patterns. This constraint instantly optimized the model's generalization capabilities, lifting the R^2 score from a critical 1.00% to 32.94% (R^2 = 0.3294).
